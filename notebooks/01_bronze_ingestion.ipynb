```python
raw_file_path = "/Volumes/suren_cat1/suren_schema/suren_volume/emp.csv"
bronze_table = "suren_cat1.suren_schema.bronze_emp"
from pyspark.sql.functions import col

# Read raw CSV from volume
df_raw = spark.read.option("header", True).csv(raw_file_path)

# Save as Bronze Delta table in Databricks
df_raw.write.mode("overwrite").format("delta").saveAsTable(bronze_table)

print(f"Data saved to Bronze table: {bronze_table}")
display(df_raw)
```
