```python
from pyspark.sql.functions import col
# Bronze table path (Delta table in volume)
bronze_table = "suren_cat1.suren_schema.bronze_emp"

# Silver table path
silver_path = "suren_cat1.suren_schema.silver_emp"

# Read Bronze table
df_bronze = spark.table(bronze_table)

# Minimal transformation: cast numeric columns, filter null names
df_silver = df_bronze \
    .withColumn("id", col("id").cast("int")) \
    .withColumn("age", col("age").cast("int")) \
    .filter(col("name").isNotNull())

# Save as Silver Delta table
df_silver.write.mode("overwrite").format("delta").saveAsTable(silver_path)

print(f"Data saved to Silver table: {silver_path}")
display(df_silver)
```
