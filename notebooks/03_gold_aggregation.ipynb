```python
gold_table = "suren_cat1.suren_schema.gold_emp"

# Silver table path
silver_table = "suren_cat1.suren_schema.silver_emp"

# Read Silver table
df_silver = spark.table(silver_table)

# Aggregation example: average age per city
df_gold = df_silver.groupBy("city").avg("age").withColumnRenamed("avg(age)", "avg_age")

# Save as Gold Delta table
df_gold.write.mode("overwrite").format("delta").saveAsTable(gold_table)

print(f"Data saved to Gold table: {gold_table}")
display(df_gold)
```
